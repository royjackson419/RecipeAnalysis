```python
#make a list of ingredients and how many times they appeared
i = 0
completeIngredientList = []

for ingredientList in df["ingredients"]:
    if isinstance(ingredientList, (list,)):
        for ingredient in ingredientList:
            try:
                    
                ingredient = ingredient.encode('ascii', 'ignore')
                print ingredient

                if ingredient in completeIngredientList.keys():
                    currentCount = completeIngredientList[ingredient]
                    completeIngredientList[ingredient] = currentCount + 1
                else:
                    completeIngredientList[ingredient] = 1
            except Exception as e:
                print (e.message)
            i += 1
            if i % 100 == 0:
                print (str(i))
      

ingredientData = pd.DataFrame(completeIngredientList.items(), columns=['Ingredient', 'TimesUsed'])

#you could save the above to a file
file_name = 'IngredientList.csv'
ingredientData.to_csv(file_name, encoding='utf-8', index=False)
```

```python
#figure out which recipes have salt

def saltQuantity(ingredientList):

        if str(type(ingredientList)) == "<type 'list'>":
            for ingredient in ingredientList:

                if ingredient.lower().endswith('salt'):
                    if ingredient.lower() not in ['kosher salt','salt']:
                        return ingredient.replace("kosher ", "").replace("fine sea ", "").replace("coarse ", "")
    


        return "None"
def hasSalt(saltInfo):

       if saltInfo == "None" or saltInfo == 'NaN':
           return 0
       else:
           return 1
    
```

```python
#populate additional fields to recipe data related to salt

df["saltQuantity"] = df["ingredients"].apply(lambda x: saltQuantity(x))

df["hasSalt"] = df["saltQuantity"].apply(lambda x: hasSalt(x))
       
```
