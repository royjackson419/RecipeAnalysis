```python
#!/usr/bin/env python
"""
Minimal Example
===============

Generating a square wordcloud from the US constitution using default arguments.
"""

from os import path
from wordcloud import WordCloud

d = os.getcwd()

# Read the whole text.
text = open(path.join(d, 'full_format_recipes.json')).read()
```

```python
import json
jdata = json.loads(text)
```

```python
ingredients = json.dumps([f["ingredients"] for f in jdata if "ingredients" in f], ensure_ascii=False).encode('utf8').replace("  "," ")
```

```python
directions = json.dumps([f["directions"] for f in jdata if "directions" in f], ensure_ascii=False).encode('utf8').replace("  "," ")
```

```python
# #!/usr/bin/env python
# """
# Minimal Example
# ===============

# Generating a square wordcloud from the US constitution using default arguments.
# """

# from os import path
# from wordcloud import WordCloud

# d = "C:\\workspace\\RecipeAnalysis\\RecipeAnalysis"

# # Read the whole text.
# text = open(path.join(d, 'full_format_recipes.json')).read()

# Generate a word cloud image
wordcloud = WordCloud().generate(directions)

# Display the generated image:
# the matplotlib way:
import matplotlib.pyplot as plt
plt.figure(figsize=(20,15))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")

# lower max_font_size
# wordcloud = WordCloud(max_font_size=40).generate(text)

# plt.imshow(wordcloud, interpolation="bilinear")
# plt.axis("off")
plt.show()

# # The pil way (if you don't have matplotlib)
# image = wordcloud.to_image()
# image.show()
```

```python

# Generate a word cloud image
wordcloud = WordCloud(width=400, height=400, collocations=False).generate(ingredients)

# Display the generated image:
# the matplotlib way:
import matplotlib.pyplot as plt
plt.figure(figsize=(20,15))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```
